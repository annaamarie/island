```python
# Create new columns for spotify followers
df_artist_metadata['sp_followers_week_before'] = df_artist_metadata['sp_followers'] - df_artist_metadata['sp_followers_diff_weekly']
df_artist_metadata['sp_followers_month_before'] = df_artist_metadata['sp_followers'] - df_artist_metadata['sp_followers_diff_monthly']

# Create new columns for spotify monthly listeners
df_artist_metadata['sp_monthly_listeners_week_before'] = df_artist_metadata['sp_monthly_listeners'] - df_artist_metadata['sp_monthly_listeners_diff_weekly']
df_artist_metadata['sp_monthly_listeners_month_before'] = df_artist_metadata['sp_monthly_listeners'] - df_artist_metadata['sp_monthly_listeners_diff_monthly']

# Create new columns for spotify popularity
df_artist_metadata['sp_popularity_week_before'] = df_artist_metadata['sp_popularity'] - df_artist_metadata['sp_popularity_diff_weekly']
df_artist_metadata['sp_popularity_month_before'] = df_artist_metadata['sp_popularity'] - df_artist_metadata['sp_popularity_diff_monthly']
```

```python
# Create dataframe for spotify time series data
time_series_data = []

# Loop through the artist metadata
for index, row in df_artist_metadata.iterrows():
    time_series_data.extend([
        {
            'artist_id': row['artist_id'],
            'date': '2024-12-13',
            'sp_followers': row['sp_followers'],
            'sp_monthly_listeners': row['sp_monthly_listeners'],
            'sp_popularity': row['sp_popularity']
        },
        {
            'artist_id': row['artist_id'],
            'date': '2024-12-06',
            'sp_followers': row['sp_followers_week_before'],
            'sp_monthly_listeners': row['sp_monthly_listeners_week_before'],
            'sp_popularity': row['sp_popularity_week_before']
        },
        {
            'artist_id': row['artist_id'],
            'date': '2024-11-13',
            'sp_followers': row['sp_followers_month_before'],
            'sp_monthly_listeners': row['sp_monthly_listeners_month_before'],
            'sp_popularity': row['sp_popularity_month_before']
        }
    ])

er_spotify_timeseries = pd.DataFrame(time_series_data)
er_spotify_timeseries.head()
```

```python
# Add column "sp_time_id"
er_spotify_timeseries = er_spotify_timeseries.assign(sp_time_id=er_spotify_timeseries.index + 1)
er_spotify_timeseries = er_spotify_timeseries[["sp_time_id", "artist_id", "date", "sp_followers", "sp_monthly_listeners", "sp_popularity"]].copy()
er_spotify_timeseries.head()
```

### Social Media Age

```python
# Instagram
social_media_age_ig = pd.melt(
    df_ig[['date', 'artist_id', 'age_13_17_male_percent', 'age_13_17_female_percent',
           'age_18_24_male_percent', 'age_18_24_female_percent',
           'age_25_34_male_percent', 'age_25_34_female_percent',
           'age_35_44_male_percent', 'age_35_44_female_percent',
           'age_45_64_male_percent', 'age_45_64_female_percent',
           'age_65_plus_male_percent', 'age_65_plus_female_percent']].assign(platform_id=1),
    id_vars=['date', 'artist_id', 'platform_id'],
    var_name='age_gender',
    value_name='percentage'
)

# TikTok
social_media_age_tt = pd.melt(
    df_tt[['date', 'artist_id', 'age_13_17_male_percent', 'age_13_17_female_percent',
           'age_18_24_male_percent', 'age_18_24_female_percent',
           'age_25_34_male_percent', 'age_25_34_female_percent',
           'age_35_44_male_percent', 'age_35_44_female_percent',
           'age_45_64_male_percent', 'age_45_64_female_percent',
           'age_65_plus_male_percent', 'age_65_plus_female_percent']].assign(platform_id=2),
    id_vars=['date', 'artist_id', 'platform_id'],
    var_name='age_gender',
    value_name='percentage'
)

# Combine Instagram and TikTok results
social_media_age = pd.concat([social_media_age_ig, social_media_age_tt], ignore_index=True)

# Extract age group and gender
social_media_age['age_group'] = social_media_age['age_gender'].str.extract(r'age_(\d{2}_\d{2}|65_plus)')
social_media_age['gender'] = social_media_age['age_gender'].str.extract(r'_(male|female)_?')[0].str.capitalize()

# Replace underscores in age group
social_media_age['age_group'] = social_media_age['age_group'].replace({
    '13_17': '13-17',
    '18_24': '18-24',
    '25_34': '25-34',
    '35_44': '35-44',
    '45_64': '45-64',
    '65_plus': '65+'
})

# Drop the original 'age_gender' column
social_media_age = social_media_age.drop(columns=['age_gender'])

# Display the final DataFrame
social_media_age.head(10)
```

```python
er_social_media_ages = social_media_age[['artist_id',
                                        'platform_id',
                                        'date',
                                        'age_group',
                                        'gender',
                                        'percentage']].copy()

# Sort by artist_id, platform_id and date
er_social_media_ages = er_social_media_ages.sort_values(['artist_id', 'platform_id', 'date']).reset_index(drop=True)

er_social_media_ages.head()
```
